```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
df.dtypes
```

1. **Count Plot**

Shows count of observation in each categorical bin using bars

```python
sns.countplot( x = 'sex' ,data = df)
```

```python
sns.countplot( y = 'day' ,data = df)
```

2. **Barplot** 

Both x axis and y axis values to be passed

```python
sns.barplot( x = 'total_bill' , y = 'sex' ,data = df)
```

In avove plot Max male bill around 20 and max Female bill around 17.5

3. **Boxplot** 

It is a graph that presents information from a 5 number summary

Evem the palette can be changed using palette parameter

```python
sns.boxplot(x = 'sex' , y = 'total_bill' , data = df)
```

```python
sns.boxplot(x = 'day' , y = 'total_bill' , data = df , palette = 'rainbow')
```

Box plot of all features are done numerical by default

```python
sns.boxplot(data = df , orient = 'v')
```

Categorize the data based on other categories like smoker mentioned via hue

```python
sns.boxplot(x = "total_bill" , y = "day" , hue = "smoker" , data = df)
```

For smoker and non smoker seperate diagrams made

4. **Violin plot** 

Violin plot helps us to see both the distribution of data in terms of Kernel density estimation and the box plot

```python
sns.violinplot(x = "total_bill" , y = "day" , data = df , palette = "rainbow")
```
