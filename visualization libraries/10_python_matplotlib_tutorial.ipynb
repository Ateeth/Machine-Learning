**MatlplotLib Tutorial**

Matplotlib is a plotting library for Pyhton programming langiage and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt or GTK+

Some major Pros of Matplotlib

- Generally easy to get started for simple plots 
- Support for custom labels and texts               
- Great control of every element in a figure
- High-quality output in many formats
- Very customizable in general

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

The reason why **%matplotlib inline** was included is we now dont need to write plot.show() to show the graph just execute the cell the graph will be made

1. **Scatter** plot using **scatter()** the c parameter is color

- g : green
- b : blue     
- r : red

**xlabel()** to specify X axis label  

**ylabel()** to specify Y axis label  

**title()** to specify graph title

**savefig()** to save the image

```python
plt.scatter(x , y , c = 'g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('images/scatter1.png')
```

```python
y = x*x
```

2. **plt** plot using **plot()** 

This is like scatter plot but the points in the plot will be connected

```python
plt.plot(x , y , 'r*--' , linewidth = 2 , markersize = 12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2D Diagram')
plt.savefig('images/plot1.png')
```

3. **Subplots** using **subplot()**

parametrs (numberOfRows , numberOfColumns , index)

```python
plt.subplot(2 , 2 , 1)
plt.plot(x , y , 'ro')

plt.subplot(2 , 2 ,2)
plt.plot(x , y , 'g')

plt.subplot(2 , 2 ,3)
plt.plot(x , y , 'b*--')

plt.savefig('images/subplot1.png')
```

Some more examples of plots

```python
x = np.arange(1 , 11)
y = 3 * x + 5
plt.title('Matplotlib demo')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.plot(x , y , 'y')
plt.savefig('images/plot2.png')
```

```python
np.pi
```

```python
x = np.arange( 0 , 4 * np.pi , 0.1)
y = np.sin(x)       
plt.title('Sine wave form')

plt.plot(x , y)
plt.savefig("images/sineplot.png")
```

```python
x = np.arange(0 , 5 * np.pi , 0.1)
y_sin = np.sin(x)  
y_cos = np.cos(x)  

##Subplot with height 2 width 1
plt.subplot(2 , 1 , 1)
plt.plot(x , y_sin , 'r--')
plt.title('Sine')

plt.subplot(2 , 1 , 2)
plt.plot(x , y_cos , 'g--')
plt.title('Cosine')

plt.savefig('images/SineAndCos.png')
```

4. Bar plot using **bar()**        

```python
x = [2 , 8 , 10]
y = [11 , 16 , 9]

x2 = [3 , 9 , 11]
y2 = [6 , 15 , 7]

plt.bar(x , y , align = "center")
plt.bar(x2 , y2 , color = "g")
plt.title("Bar graph")
plt.ylabel("Y axis")
plt.xlabel("X axis")

plt.savefig("images/bar1.png")
```

Histogram using **hist()**

```python
a = np.array([22 , 87 , 5 , 43 , 56 , 73 , 54 , 55 , 11 , 20 , 51 , 5 , 79 , 31 , 27])
```

**bins** to specify how many bars it will be split into

```python
plt.hist(a , bins = 10)
plt.title("Histogram")
plt.savefig("images/histogram1.png")
```

5. Boxplot using **boxplot()**

- vertical : vertical box alignment
- patch_artist : fill with color

```python
data = [np.random.normal(0 , std , 100) for std in range(1 , 4)]

plt.boxplot(data, vert = True , patch_artist= True)
plt.savefig("images/boxplot1.png")
```

6. Pie chart using **pie()**                         

- autopct = indicates in what way the amount of value in the graph occupied is shown 

In the below snippet it is shown in floating point

```python
labels = ['Python' , 'C++' , 'Ruby' , 'Java']

sizes = [215 , 130 , 245 , 210]

colors = ['gold' , 'yellowgreen' , 'lightcoral' , 'lightskyblue']

##Explode the first slice
explode = (0.1 , 0 , 0, 0)

plt.pie(sizes , explode = explode , labels = labels , colors = colors , autopct = '%1.1f%%' , shadow = True)

plt.axis('Equal')
plt.savefig("images/pie1.png")
```
